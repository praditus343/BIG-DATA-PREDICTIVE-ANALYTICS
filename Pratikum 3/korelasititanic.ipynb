```python
# memuat numpy sebagai np
import numpy as np
 
# memuat pandas sebagai pd
import pandas as pd

import seaborn as sns
```

```python
# memuat data bernama 'dataset_statistics.csv' dan memasukkan hasilnya ke dalam 'raw_data'
raw_data = pd.read_csv('https://raw.githubusercontent.com/muliasulistiyono/ST152-Big-Data-Predictive-Analysis/main/Materi%20Praktikum/Pertemuan%203/titanic%20-%20titanic.csv')
```

```python
raw_data.info()
```

```python
print (raw_data)
# melihat 10 data pada baris pertama
print (raw_data.head(10))

```

```python
# melihat 5 data pada baris terakhir
print (raw_data.tail())
```

```python
#Kolom yang ada pada dataset
raw_data.drop(columns=['Name','Sex','Ticket','Ticket','Embarked','Cabin'], inplace=True)
```

```python
# menghitung korelasi dari setiap pasang variabel pada raw_data
print (raw_data.corr())
```

```python
# mencari korelasi 'kendall' untuk tiap pasang variabel
print (raw_data.corr(method='kendall'))
```

```python
# mencari korelasi 'spearman' untuk tiap pasang variabel
print (raw_data.corr(method='spearman'))
```

```python
#melihat nilai korelasi antar variabel Jenis Kelamin dan Pendapatan
raw_data['PassengerId'].corr(raw_data['Age'], method='spearman')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.clf()

# mengatur ukuran gambar/plot
plt.rcParams['figure.dpi'] = 100

plt.figure()
plt.matshow(raw_data.corr())
plt.title('Plot correlation matriks dengan .matshow', size=14)
plt.tight_layout()
plt.show()

plt.figure()
sns.heatmap(raw_data.corr(), annot=True)
plt.title('Plot correlation matriks dengan sns.heatmap', size=14)
plt.tight_layout()
plt.show()
```

# Grouped Box Plot

```python
import matplotlib.pyplot as plt
plt.clf()

plt.figure()
# boxplot biasa tanpa pengelompokkan
raw_data.boxplot(rot=90)
plt.title('Boxplot tanpa pengelompokkan', size=14)
plt.tight_layout()
plt.show()

plt.figure()
# box plot dengan pengelompokkan dilakukan oleh kolom 'Age'
raw_data.boxplot(by='Fare')
plt.tight_layout()
plt.show()
```

# Grouped Histogram

```python
import matplotlib.pyplot as plt
plt.clf()

plt.figure()
raw_data[raw_data['Fare'] == 'A'].hist()
plt.tight_layout()
plt.show()

plt.figure()
raw_data[raw_data['Fare'] == 'B'].hist()
plt.tight_layout()
plt.show()

plt.figure()
raw_data[raw_data['Fare'] == 'C'].hist()
plt.tight_layout()
plt.show()

plt.figure()
raw_data[raw_data['Fare'] == 'D'].hist()
plt.tight_layout()
plt.show()

plt.figure()
raw_data[raw_data['Fare'] == 'E'].hist()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.clf()

plt.figure()
raw_data.plot.hexbin(x='Age', y='Fare', gridsize=25, rot=90)
plt.tight_layout()
plt.show()
```

# Scatter Matrix Plot

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
plt.clf()

_, ax = plt.subplots(1, 1, figsize=(10,10))
scatter_matrix(raw_data, ax=ax)
plt.show()
```
